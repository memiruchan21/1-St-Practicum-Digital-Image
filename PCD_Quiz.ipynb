```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# bgr
img = cv2.imread('/content/81f3b9594359687b5d2aca19013943b7.jpg')

#rgb
Lizzie = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Showing a picture without filter
cv2_imshow(img)

# Making a filter; Matrix 5 x 5
kernel = np.ones((5,5),np.float32)/25
print(kernel)

# Filtering
Liz_filter = cv2.filter2D(img,-1,kernel)

cv2_imshow(Liz_filter)
```

```python
# Making size plotting bigger (if it's necessary)
plt.rcParams["figure.figsize"] = (15,15)

#first plot, real pict
plt.subplot(121),plt.imshow(Lizzie),plt.title('Original')
plt.xticks([]), plt.yticks([])

#second, filter result
plt.subplot(122),plt.imshow(Liz_filter),
plt.title('Averaging')
plt.xticks([]), plt.yticks([])

#Plot!
plt.show()
```

```python
Liz_blur = cv2.blur(img,(5,5))

cv2_imshow(Liz_blur)
```

```python
# This is another way to make some kernel
# with using np.matrix
# This time, matrix size is 3 x 3

kernel = np.matrix([
                    [1, 1, 1],
                    [1, 2, 1],
                    [1, 1, 1]
])/25

print(kernel)

#Making a new filter
Liz_filter = cv2.filter2D(img,-1,kernel)

#Show
cv2_imshow(Liz_filter)
```
