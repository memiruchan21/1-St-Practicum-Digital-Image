```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python

lego_a = cv2.imread('/kaggle/input/bingung/1656822203745.png',1)
lego_a = cv2.cvtColor(lego_a, cv2.COLOR_BGR2RGB)

lego_b = cv2.imread('/kaggle/input/huhuhuhuhu/1656822376589.png',1)
lego_b = cv2.cvtColor(lego_b, cv2.COLOR_BGR2RGB)
```

```python
lego_blending = cv2.addWeighted(lego_a, 0.7, lego_b, 0.5,0)
plt.figure(figsize=(20, 20))
plt.subplot(1,3,1)
plt.imshow(lego_a)
plt.subplot(1,3,2)
plt.imshow(lego_b)
plt.subplot(1,3,3)
plt.imshow(lego_blending)
plt.show()
```

```python

img = cv2.imread('/kaggle/input/lizzie/81f3b9594359687b5d2aca19013943b7.jpg',0)
img_negative = 255 - img
plt.figure(figsize=(15, 15))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_negative,cmap="gray")
plt.show()
```

```python
from PIL import Image, ImageEnhance
#read the image
im = Image.open("/kaggle/input/huhuhuhuhu/1656822376589.png")
#image brightness enhancer
enhancer = ImageEnhance.Brightness(im)
factor = 1 #gives original image
im_output = enhancer.enhance(factor)
im_output.save('asli.png')
factor = 0.5 #darkens the image
im_output = enhancer.enhance(factor)
im_output.save('gelap.png')
factor = 2.5 #brightens the image
im_output = enhancer.enhance(factor)
im_output.save('cerah.png')
plt.show()
```

```python
img = cv2.imread('/kaggle/input/bingung/1656829968379.jpg',0)
gamma = 2.5
img_power = img**gamma
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_power,cmap="gray")
plt.show()
```

```python
img = cv2.imread('/kaggle/input/lizzie/81f3b9594359687b5d2aca19013943b7.jpg',1)
kernel = np.ones((3, 3), np.float32) / 9
img_out = cv2.filter2D(img, -1, kernel)
plt.figure(figsize=(20, 20))
plt.subplot(1,2,2)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_out,cmap="gray")
plt.show()
```
